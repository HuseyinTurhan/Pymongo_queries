```python
import pymongo
import json
myclient = pymongo.MongoClient(host='mongodb://localhost:27017')

mydb = myclient['PythonMongoDb']
print("Databases available in the MongoDB are: ", myclient.list_database_names())

mycollection = mydb['Restaurants']
mydict = {  "name": "New Stockholm Restaurant",
            "borough": "Stockholm",
            "restaurant_id": "30042574"}
mycollection.insert_one(mydict)

print("Collections available in the PythonMongoDb database are: ", mydb.list_collection_names())
```

```python
mycol_json = mydb['Restaurants_json']
# Loading or Opening the json file
file_data = [json.loads(line) for line in open('restaurants.json', 'r')]

# Inserting the loaded data in the Collection
# if JSON contains data more than one entry
# insert_many is used else insert_one is used
if isinstance(file_data, list):
    mycol_json.insert_many(file_data) 
else:
    mycol_json.insert_one(file_data)

print("Collections available in the PythonMongoDb database are: ", mydb.list_collection_names())
```

```python
print("Documents from manuel entered Restaurants collection:")
for z in mycollection.find():
    print(z)

print("A few documents from Restaurants collection which is uploaded from json file:")
for z in mycol_json.find().limit(3):
    print(z)
```

```python
# Problem-2 and Problem-3
# Write a MongoDB query to display the fields restaurant_id, name, borough and cuisine for all the documents in the collection restaurant
for z in mycol_json.find({},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-5
# Write a MongoDB query to display all the restaurant which is in the borough Bronx.
for z in mycol_json.find({"borough":"Bronx"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-6
# Write a MongoDB query to find the restaurants who achieved a score more than 90.

for z in mycol_json.find({"grades.score":{"$gt":90}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-7
# Write a MongoDB query to find the restaurants that achieved a score, more than 80 but less than 100.

for z in mycol_json.find({"grades":{"$elemMatch":{"score" :{"$gt":80,"$lt":100 }}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-8
# Write a MongoDB query to find the restaurants which locate in latitude value less than -95.754168.

for z in mycol_json.find({"address.coord" :{"$lt":-95.754168}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-9 and Problem-10
# Write a MongoDB query to find the restaurants that do not prepare any cuisine of 'American' and their grade score more than 70 and latitude less than -65.754168.

for z in mycol_json.find({"address.coord" :{"$lt":-65.754168}, "cuisine":{"$not":{"$regex":"^American*"}}, "grades":{"$elemMatch":{"score" :{"$gt":70}}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-11
# Write a MongoDB query to find the restaurants which belong to the borough Bronx and prepared either American or Chinese dish.

for z in mycol_json.find({"borough":"Bronx", "$or":[{"cuisine":{"$regex":"^American*"}},{"cuisine":{"$regex":"^Chinese*"}}]},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-12
# Write a MongoDB query to find the restaurant Id, name, borough and cuisine for those restaurants which belong to the borough Staten Island or Queens or Bronxor Brooklyn.

for z in mycol_json.find({"borough":{"$in":["Staten Island", "Queens","Bronx","Brooklyn"]}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-13
# Write a MongoDB query to find the restaurant Id, name, borough and cuisine for those restaurants which are not belonging to the borough Staten Island or Queens or Bronxor Brooklyn.

for z in mycol_json.find({"borough":{"$not":{"$in":["Staten Island", "Queens","Bronx","Brooklyn"]}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# Problem-14
# Write a MongoDB query to find the restaurant Id, name, borough and cuisine for those restaurants which achieved a score which is not more than 10.

for z in mycol_json.find({"grades":{"$not":{"$elemMatch":{"score":{"$gt":10}}}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print(z)
```

```python
# updating a document (update_one)
for z in mycol_json.find({"restaurant_id":"40367179"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("Restaurant informations before updating document:\n", z)

result = mycol_json.update_one({"restaurant_id":"40367179"}, {"$set":{"name":"Stockholm Lexicon Python"}}, upsert=True)
print("UpdateResult( nMatched : ", result.matched_count, ", nModified : ", result.modified_count, ", nUpsertedId : ", result.upserted_id, " )")
for z in mycol_json.find({"restaurant_id":"40367179"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("After updating document:\n", z)
```

```python
# updating a few document (update_many)
for z in mycol_json.find({"cuisine":"Irish"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("Restaurant informations before updating document:\n", z)

result = mycol_json.update_many({"cuisine":"Irish"}, {"$set":{"name":"Irish Restaurant"}}, upsert=True)
print("UpdateResult( nMatched : ", result.matched_count, ", nModified : ", result.modified_count, ", nUpsertedId : ", result.upserted_id, " )")
for z in mycol_json.find({"cuisine":"Irish"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("After updating document:\n", z)
```

```python
# deleting one document (delete_one)
for z in mycol_json.find({"name":"Stockholm Lexicon Python"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("Restaurant informations before updating document:\n", z)

mycol_json.delete_one({"name":"Stockholm Lexicon Python"})
for z in mycol_json.find({"name":"Stockholm Lexicon Python"},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("After updating document:\n", z)
```

```python
# deleting a few document (delete_many)
for z in mycol_json.find({"grades":{"$elemMatch":{"score":{"$lt":10}}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("Restaurant informations before updating document:\n", z)


d = mycol_json.delete_many({"grades":{"$elemMatch":{"score":{"$lt":10}}}})
print(d.deleted_count, " documents deleted !!")
for z in mycol_json.find({"grades":{"$elemMatch":{"score":{"$lt":10}}}},{"_id":0,"restaurant_id":1,"name":1,"borough":1, "cuisine":1}):
    print("After updating document:\n", z)
```
